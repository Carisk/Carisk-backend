```python
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow import keras
from sklearn.preprocessing import OneHotEncoder
```

```python
data = pd.read_csv('data_pre.csv', encoding="UTF-8")
```

```python
df = pd.DataFrame(data=data)
```

```python
df.head()
```

```python
df = df[['dia_semana', 'uf', 'fase_dia', 'condicao_metereologica', 'latitude', 'longitude', 'tipo_acidente']]
```

```python
df.head()
```

```python
len(df)
```

```python
train, test = train_test_split(df, test_size=0.2, random_state=42, shuffle=True)
```

```python
train
```

```python
train_label = train[['tipo_acidente']]
```

```python
train_label
```

```python
encoder = OneHotEncoder(handle_unknown='ignore')
train_label = pd.DataFrame(encoder.fit_transform(train_label[['tipo_acidente']]).toarray())
train_label.head()
```

```python
train = train[['dia_semana', 'uf', 'fase_dia', 'condicao_metereologica', 'latitude', 'longitude']]
train
```

```python
test
```

```python
test_label = test[['tipo_acidente']]
test_label
```

```python
encoder = OneHotEncoder(handle_unknown='ignore')
test_label = pd.DataFrame(encoder.fit_transform(test_label[['tipo_acidente']]).toarray())
test_label.head()
```

```python
test = test[['dia_semana', 'uf', 'fase_dia', 'condicao_metereologica', 'latitude', 'longitude']]
test
```

```python
from tensorflow.keras import layers
```

```python
model = keras.Sequential()
model.add(layers.Dense(6, activation="relu", input_shape=(train.shape[-1],)))
model.add(layers.Dense(256, activation="relu"))
model.add(layers.Dropout(0.1))
model.add(layers.Dense(256, activation="relu"))
model.add(layers.Dropout(0.1))
model.add(layers.Dense(8, activation="sigmoid"))        
```

```python
model.summary()
```

```python
metrics = [
    keras.metrics.Accuracy(name="accuracy")    
]

model.compile(
    optimizer=keras.optimizers.Adam(1e-2), loss="categorical_crossentropy", metrics=metrics
)
```

```python
model.fit(
    train,
    train_label,
    batch_size=128,
    epochs=10,
    verbose=0,
    validation_data=(test, test_label),
)
```
